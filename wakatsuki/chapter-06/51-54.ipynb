```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#51
#各ファイル読み込んでリスト化

#FORMAT: ID \t TITLE \t URL \t PUBLISHER \t CATEGORY \t STORY \t HOSTNAME \t TIMESTAMP

test_lines_list = []
valid_lines_list = []
train_lines_list = []


with open("test.txt") as f1:
    for line in f1.readlines():
        test_lines_list.append(line.strip().split('\t')[1])
with open("valid.txt") as f2:
    for line in f2.readlines():
        valid_lines_list.append(line.strip().split('\t')[1])
with open("train.txt") as f3:
    for line in f3.readlines():
        train_lines_list.append(line.strip().split('\t')[1])

print(test_lines_list[0:3])
print(valid_lines_list[0:3])
print(train_lines_list[0:3])

print(len(test_lines_list))
print(len(valid_lines_list))
print(len(train_lines_list))

doc_lines_list = []
print(type(doc_lines_list))
doc_lines_list.extend(test_lines_list)
doc_lines_list.extend(valid_lines_list)
doc_lines_list.extend(train_lines_list)

print(len(doc_lines_list))
```

```python

# CountVectorizer 
c_vectorizer = CountVectorizer(binary=True)

# ベクトル化 
c_vectorizer.fit(doc_lines_list)
print(type(c_vectorizer))

#特徴量の計算
X_test = c_vectorizer.transform(test_lines_list)
X_valid = c_vectorizer.transform(valid_lines_list)
X_train = c_vectorizer.transform(train_lines_list)

# print(type(X_test))
print(X_test)

#抽出した単語と出現回数の辞書
voc_count_dict = c_vectorizer.vocabulary_
# print(type(voc_count_dict))
# print(len(voc_count_dict))
# print(voc_count_dict)

#抽出単語/各行　値：出現したかのバイナリ
pd.DataFrame(X_test.toarray(), columns=c_vectorizer.get_feature_names())
# test_feature = X_test.toarray()
# print(test_feature)
```

```python
#正解ラベル生成
Y_test = []
Y_valid = []
Y_train = []

with open("test.txt") as f1:
    for line in f1.readlines():
        Y_test.append(line.strip().split('\t')[4])
with open("valid.txt") as f2:
    for line in f2.readlines():
        Y_valid.append(line.strip().split('\t')[4])
with open("train.txt") as f3:
    for line in f3.readlines():
        Y_train.append(line.strip().split('\t')[4])

print(len(Y_test))
print(Y_test)
print(len(Y_valid))
print(Y_valid)
print(len(Y_train))
print(Y_train)
```

```python
#52 CountVectorizer

lr = LogisticRegression() # ロジスティック回帰モデルのインスタンスを作成
lr.fit(X_train, Y_train) # ロジスティック回帰モデルの重みを学習

regression_coefficient = lr.coef_
segment = lr.intercept_
print(len(lr.coef_[1]))
print("coefficient = ", lr.coef_)
print("intercept = ", lr.intercept_)
print("決定係数:{}".format(lr.score(X_train, Y_train)))

```

```python
#53 予測確率
probs = lr.predict_proba(X_test)
print(probs)
```

```python
#54 正答率
Y_pred = lr.predict(X_test)
print(len(Y_pred))
print(type(Y_pred))
print(Y_pred)
print(len(Y_pred == Y_test))
print(type(Y_pred == Y_test))
print(Y_pred == Y_test)
correct_answer = np.count_nonzero(Y_pred == Y_test)

print('正解率 : ' + str(correct_answer/ len(Y_pred)))
```

```python

# TfidfVectorizer
t_vectorizer = TfidfVectorizer()


print(test_lines_list[0:3])
print(test_lines_list[0:3])

# ベクトル化 
t_vectorizer.fit(doc_lines_list)
print(type(t_vectorizer))

#特徴量の計算
X_test_t = t_vectorizer.transform(test_lines_list)
X_valid_t = t_vectorizer.transform(valid_lines_list)
X_train_t = t_vectorizer.transform(train_lines_list)

# print(type(X_test))
print(X_test_t)

#抽出した単語と出現回数の辞書
# tiidf_dict = t_vectorizer.vocabulary_
# print(type(tiidf_dict))
# print(len(tiidf_dict))
# print(tiidf_dict)

#抽出単語/各行　値：出現したかのバイナリ
pd.DataFrame(X_test_t.toarray(), columns=t_vectorizer.get_feature_names())
# test_feature = X_test.toarray()
# print(test_feature)
```

```python
#53 予測確率 TfidfVectorizer
probs = lr.predict_proba(X_test_t)
print(probs)
```
